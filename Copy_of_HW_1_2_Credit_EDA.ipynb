##Аналіз колонки EXT_SOURCE_2

Спочатку створюємо функцію, яка б виводила пропущені значення (кількість і відсоток) і значення стовпця. потім йде побудова боксплот графіка без пропущених значень і виведення значень квантилів, далі мода, середнє і медіана

```python
def analyze_column_under1(df, col):
  c = df[col]
  null_count = int(c.isna().sum())
  null_percentage = float(c.isna().mean() * 100)
  print(f"null_count: {null_count}")
  print(f"null_percentage: {null_percentage:.2f}%")
  vc = c.value_counts().head(10)
  print(vc)
  sns.boxplot(y=c.dropna())
  plt.title(c.name)
  plt.show()
  q = c.dropna().quantile([0.25, 0.5, 0.75, 1.0])
  print(q)
  mode_val = c.dropna().mode().iloc[0]
  mean_val = float(c.mean())
  median_val = float(c.median())
  print(f"\nMode: {mode_val}")
  print(f"Mean: {mean_val}")
  print(f"Median: {median_val}")
```

```python
result = analyze_column_under1(df, "EXT_SOURCE_2")
```

Тут ми бачимо, що відсоток пропусків досить малий, а також, що медіана і середнє трохи відрізняються, тобто розподіл скошений. Тому вважаю, що пропущені значення варто заповнити медіаною, оскільки вона стійка до скошеного розподілу.

```python
df["EXT_SOURCE_2"].fillna(df["EXT_SOURCE_2"].median(), inplace=True)
```

### Створення функції `age_cat` для категоризації `YEARS_BORNING`

```python
def age_cat(years):
    if years <= 20:
        return '0-20'
    elif years > 20 and years <= 30:
        return '20-30'
    elif years > 30 and years <= 40:
        return '30-40'
    elif years > 40 and years <= 50:
        return '40-50'
    elif years > 50 and years <= 60:
        return '50-60'
    elif years > 60 and years <= 70:
        return '60-70'
    elif years > 70:
        return '70+'
```

### Одномірний (категоріальний vs цільовий) та двомірний (категоріальний vs категоріальний) аналіз (гістограма) - категоріальні стовпчики

```python
import matplotlib.pyplot as plt


def bi_cat_countplot(df, column, hue_column):
    unique_hue_values = df[hue_column].unique()
    fig, axes = plt.subplots(nrows=1, ncols=2)
    fig.set_size_inches(14,6)

    pltname = f'Нормалізований розподіл значень за категорією: {column}'
    proportions = df.groupby(hue_column)[column].value_counts(normalize=True)
    proportions = (proportions*100).round(2)
    ax = proportions.unstack(hue_column).sort_values(
        by=unique_hue_values[0], ascending=False
        ).plot.bar(ax=axes[0], title=pltname)

    # анотація значень в барплоті
    for container in ax.containers:
        ax.bar_label(container, fmt='{:,.1f}%')

    pltname = f'Кількість даних за категорією: {column}'
    counts = df.groupby(hue_column)[column].value_counts()
    ax = counts.unstack(hue_column).sort_values(
        by=unique_hue_values[0], ascending=False
        ).plot.bar(ax=axes[1], title=pltname)

    for container in ax.containers:
      ax.bar_label(container)


def uni_cat_target_compare(df, column):
    bi_cat_countplot(df, column, hue_column='TARGET' )


def bi_countplot_target(df0, df1, column, hue_column):
  pltname = 'Клієнт зі складнощами щодо платності'
  print(pltname.upper())
  bi_cat_countplot(df1, column, hue_column)
  plt.show()

  pltname = 'Клієнти зі своєчасними платежами'
  print(pltname.upper())
  bi_cat_countplot(df0, column, hue_column)
  plt.show()
```

### Обчислення мінімального та максимального діапазону відхилень для числових стовпців

```python
def outlier_range(dataset,column):
    Q1 = dataset[column].quantile(0.25)
    Q3 = dataset[column].quantile(0.75)
    IQR = Q3 - Q1
    Min_value = (Q1 - 1.5 * IQR)
    Max_value = (Q3 + 1.5 * IQR)
    return Max_value
```

##Одновимірний аналіз колонок YEARS_BIRTH, AMT_GOODS_PRICE, DAYS_EMPLOYED

```python


def one_analysis(df, column, group_col, group1_val, group2_val):

    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    min_value = Q1 - 1.5 * IQR
    max_value = Q3 + 1.5 * IQR

    print(f"IQR: {IQR:.2f}")
    print(f"Мінімальна межа: {min_value:.2f}")
    print(f"Максимальна межа: {max_value:.2f}")


    df_group1 = df[df[group_col] == group1_val]
    df_group2 = df[df[group_col] == group2_val]

    df_group1_no = df_group1[(df_group1[column] >= min_value) & (df_group1[column] <= max_value)]
    df_group2_no = df_group2[(df_group2[column] >= min_value) & (df_group2[column] <= max_value)]


    plt.figure(figsize=(14,6))
    sns.kdeplot(df_group1_no[column], label=f"{group_col}={group1_val}", linewidth=2)
    sns.kdeplot(df_group2_no[column], label=f"{group_col}={group2_val}", linewidth=2)
    plt.title(f"Розподіл {column} після видалення викидів")
    plt.ylabel("Density")
    plt.legend()
    plt.show()

    return {
        'IQR': IQR,
        'Min': min_value,
        'Max': max_value,
        'Group1_no_outliers': df_group1_no,
        'Group2_no_outliers': df_group2_no
    }
```

```python
results = one_analysis(df, 'YEARS_BIRTH', 'TARGET', 1, 0)
```

Серед вікової групи до 40 років ми бачимо що переважають заявники, які мають пролеми з поверненням кредитів, тоді як після 40 років - навпаки переважають добросовісні позичальники.

```python
results = one_analysis(df, 'AMT_GOODS_PRICE', 'TARGET', 1, 0)
```

Тут ми не бачимо стійкої тенденції у впливі ціни товару на вчасність оплат, лише на піках найчастіших цін ми бачимо, що чим частіше зустрічається цінове значення, ти добросовішніщі позичальники.

##Кореляційний аналіз пари AMT_ANNUITY та AMT_CREDIT

```python
def bivariate(df, col_x, col_y,
                              target_col='TARGET',
                              group1_val=1, group0_val=0):

    def outlier_max(dataset, column):
        Q1 = dataset[column].quantile(0.25)
        Q3 = dataset[column].quantile(0.75)
        IQR = Q3 - Q1
        max_value = Q3 + 1.5 * IQR
        return max_value


    df1 = df[df[target_col] == group1_val]
    df0 = df[df[target_col] == group0_val]


    max1_x = outlier_max(df1, col_x)
    max1_y = outlier_max(df1, col_y)
    max0_x = outlier_max(df0, col_x)
    max0_y = outlier_max(df0, col_y)

    df1_clean = df1[(df1[col_x] < max1_x) & (df1[col_y] < max1_y)]
    df0_clean = df0[(df0[col_x] < max0_x) & (df0[col_y] < max0_y)]


    plt.figure(figsize=(14, 6))


    plt.subplot(1, 2, 1)
    plt.title('Payment difficulties')
    sns.scatterplot(x=df1_clean[col_x], y=df1_clean[col_y])
    plt.xlabel(col_x)
    plt.ylabel(col_y)
    plt.ticklabel_format(style='plain', axis='x')
    plt.ticklabel_format(style='plain', axis='y')


    plt.subplot(1, 2, 2)
    plt.title('On-Time Payments')
    sns.scatterplot(x=df0_clean[col_x], y=df0_clean[col_y])
    plt.xlabel(col_x)
    plt.ylabel(col_y)
    plt.ticklabel_format(style='plain', axis='x')
    plt.ticklabel_format(style='plain', axis='y')

    plt.tight_layout(pad=4)
    plt.show()


```

```python
bivariate(df,
                          col_x='AMT_ANNUITY',
                          col_y='AMT_CREDIT',
                          target_col='TARGET',
                          group1_val=1,
                          group0_val=0)
```

На даних графіках ми бачимо чітку позитивну кореляцію між сумою кредиту та щомісячними платежами. Той факт, що зі зростанням суми кредиту, графік ніби розсіюється, пояснюється технічними причинами, а саме відсотком по кредиту і терміном позики, а не патернами в поведінці заявника. Чим вищий відсоток по кредиту і коротший термін кредиту, тим вище буде щомісячний платіж. І навпаки - чим нижчий відсоток і довший термін позики, тим меншим буде платіж при тій самій сумі кредиту.
  Одним словом, причина віялоподібності графіка в наперед визначених умовах договору.

## Неперервні та категоріальні змінні

```python
def draw_boxplot(df, categorical, continuous, max_continuous, title, hue_column, subplot_position):

    plt.subplot(1, 2, subplot_position)
    plt.title(title)
    red_diamond = dict(markerfacecolor='r', marker='D')
    sns.boxplot(x=categorical,
                y=df[df[continuous] < max_continuous][continuous],
                data=df,
                flierprops=red_diamond,
                order=sorted(df[categorical].unique(), reverse=True),
                hue=hue_column, hue_order=sorted(df[hue_column].unique(), reverse=True))
    plt.ticklabel_format(style='plain', axis='y')
    plt.xticks(rotation=90)

def bi_boxplot(categorical, continuous, max_continuous1, max_continuous0, hue_column):

    plt.figure(figsize=(16, 10))

    draw_boxplot(df1, categorical, continuous, max_continuous1, 'Payment Difficulties', hue_column, 1)

    draw_boxplot(df0, categorical, continuous, max_continuous0, 'On-Time Payments', hue_column, 2)

    plt.tight_layout(pad=4)
    plt.show()
```

Зробіть висновок про наявність кореляції між змінними.


```python
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import display

# Допоміжна: верхня межа викидів за правилом 1.5*IQR
def outlier_range(dataset, column):
    Q1 = dataset[column].quantile(0.25)
    Q3 = dataset[column].quantile(0.75)
    IQR = Q3 - Q1
    max_value = Q3 + 1.5 * IQR
    return max_value

def triple_cat_num_analysis(df,
                            num_col,
                            cat_col1,
                            cat_col2,
                            target_col='TARGET',
                            bad_val=1,
                            good_val=0):

    df_bad = df[df[target_col] == bad_val]
    df_good = df[df[target_col] == good_val]

    max_bad = outlier_range(df_bad, num_col)
    max_good = outlier_range(df_good, num_col)

    print(f"Максимальне значення {num_col} без викидів для групи з платіжними труднощами: {max_bad}")
    print(f"Максимальне значення {num_col} без викидів для групи з вчасними платежами: {max_good}\n")

    print("Клієнти з платіжними труднощами:")
    grp_bad = (df_bad
               .groupby([cat_col1, cat_col2])[num_col]
               .agg(['count', 'mean', 'std']))
    display(grp_bad)

    print("\nКлієнти з вчасними платежами:")
    grp_good = (df_good
                .groupby([cat_col1, cat_col2])[num_col]
                .agg(['count', 'mean', 'std']))
    display(grp_good)

    plt.figure(figsize=(16, 6))

    plt.subplot(1, 2, 1)
    plt.title('Payment difficulties')
    sns.boxplot(
        data=df_bad[df_bad[num_col] < max_bad],
        x=cat_col1,
        y=num_col,
        hue=cat_col2
    )
    plt.xticks(rotation=45)
    plt.ticklabel_format(style='plain', axis='y')

    plt.subplot(1, 2, 2)
    plt.title('On-Time Payments')
    sns.boxplot(
        data=df_good[df_good[num_col] < max_good],
        x=cat_col1,
        y=num_col,
        hue=cat_col2
    )
    plt.xticks(rotation=45)
    plt.ticklabel_format(style='plain', axis='y')

    plt.tight_layout(pad=4)
    plt.show()


```

```python
triple_cat_num_analysis(
    df,
    num_col='AMT_INCOME_TOTAL',
    cat_col1='NAME_FAMILY_STATUS',
    cat_col2='CODE_GENDER',
    target_col='TARGET'
)
```

На правому ми бачимо, що жінки, незалежно від сімейного статусу, навіть маючи менший дохід, чим чоловіки, схильні сумлінно сплачувати позики.
